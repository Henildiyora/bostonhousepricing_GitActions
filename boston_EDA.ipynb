```python
## deviding the data into train and test
from sklearn.model_selection import train_test_split

X = df.drop('price',axis=1)
y = df['price']

X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,random_state=42)

```

```python
## Standardizing the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
# Model training 
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score
import optuna


def adjusted_r2(r2_score,y_test,X_test):

    adjusted_r2 = 1 - (1-r2_score)*(len(y_test)-1)/(len(y_test) - X_test.shape[1]-1)

    return adjusted_r2


def objective(trial):

    model = trial.suggest_categorical('model', ['LinearRegression', 'DecisionTreeRegressor', 'RandomForestRegressor', 'SVR'])

    if model == 'LinearRegression':
        model_params = {}

    elif model == 'DecisionTreeRegressor':
        model_params = {
            'criterion': trial.suggest_categorical('criterion', ['friedman_mse', 'squared_error', 'absolute_error', 'poisson']),
            'max_depth': trial.suggest_int('max_depth', 2, 32),
            'min_samples_split': trial.suggest_int('min_samples_split', 2, 32),
            'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 32)
        }

    elif model == 'RandomForestRegressor':
       model_params = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 300),
        'criterion': trial.suggest_categorical('criterion', ['friedman_mse', 'squared_error', 'absolute_error', 'poisson']),
        'max_depth': trial.suggest_int('max_depth', 2, 32),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 32),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 32)
        }

    elif model == 'SVR':
        model_params = {
            'kernel': trial.suggest_categorical('kernel', ['linear', 'poly', 'rbf']),
            'C': trial.suggest_int('C', 1, 100),
            'gamma': trial.suggest_categorical('gamma', ['scale', 'auto']),
            'epsilon': trial.suggest_int('epsilon', 1, 100)
        }

    # Instantiate the model with the suggested parameters
    if model == 'LinearRegression':
        regressor = LinearRegression(**model_params)
    elif model == 'DecisionTreeRegressor':
        regressor = DecisionTreeRegressor(**model_params)
    elif model == 'RandomForestRegressor':
        regressor = RandomForestRegressor(**model_params)
    elif model == 'SVR':
        regressor = SVR(**model_params)

    

    # Fit the model
    regressor.fit(X_train, y_train)

    pred = regressor.predict(X_test)

    adj_r2  = adjusted_r2(y_test=y_test,r2_score=r2_score(y_test,pred),X_test=X_test)

    return adj_r2


study = optuna.create_study(direction='maximize')
study.optimize(objective,n_trials=200)


parms = study.best_params

model = parms['model']
del parms['model']

model_params = parms

print(model)
print(model_params)


if model == 'LinearRegression':
    regressor = LinearRegression(**model_params)
elif model == 'DecisionTreeRegressor':
    regressor = DecisionTreeRegressor(**model_params)
elif model == 'RandomForestRegressor':
    regressor = RandomForestRegressor(**model_params)
elif model == 'SVR':
    regressor = SVR(**model_params)

regressor.fit(X_train, y_train)

new_pred = regressor.predict(X_test)

new_adj_r2  = adjusted_r2(y_test=y_test,r2_score=r2_score(y_test,new_pred),X_test=X_test)

print(new_adj_r2)




```
